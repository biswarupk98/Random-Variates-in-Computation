```python
import numpy as np
from scipy.stats import truncnorm
from scipy.stats import uniform
from scipy.stats import arcsine
from scipy.stats import beta
import matplotlib.pyplot as plt
from scipy import stats
a = -3                        # left end value for truncated normal
b = 3                         # Right end value for truncated normal
```


#CDF for Truncated Normal Distribution

```python
N_ksT=[]
U_ksT=[]
A_ksT=[]
def Confusion_T(samples,trial):
  np.random.seed(10)
  x=truncnorm.rvs(a,b,loc=0.5,scale=1/6,size=samples)
  x.sort()
  N_cdf=truncnorm.cdf(x,a,b,loc=0.5,scale=1/6)
  U_cdf=uniform.cdf(x,0,1)
  A_cdf=beta.cdf(x,0.5,0.5,loc=0,scale=1)
  #Empirical Distribution Function
  edf_=(np.arange(1,len(x)+1))/len(x)
  #KS Values   
  N_ksT.append(np.max(np.abs(N_cdf-edf_)))    
  U_ksT.append(np.max(np.abs(U_cdf-edf_)))    
  A_ksT.append(np.max(np.abs(A_cdf-edf_)))
  # PLOT
  plt.plot(x, N_cdf ,  label='truncated normal cdf ',color='blue') 
  plt.plot(x,U_cdf, label='Uniform cdf',color='red')
  plt.plot(x,A_cdf, label='Arcsine cdf',color='green')     
  plt.plot(x, edf_ ,  label='Edf ',color='black')   
  plt.legend()
  plt.title(("CDFs and EDF for n= " , samples))  
  plt.show() 
  

 
```

```python
plt.figure(figsize=(5, 5))
for trial in range (0,3):
  for i in range (2,6):
    Confusion_T(10**i,trial)
    

print(len(N_ksT))
print(N_ksT)
```


#Arcsine Distribution

```python
N_ksA=[]
U_ksA=[]
A_ksA=[]
def Confusion_A(samples,trial):
  np.random.seed(10)
  x=beta.rvs(0.5,0.5,size=samples)
  x.sort()
  N_cdf=truncnorm.cdf(x,a,b,loc=0.5,scale=1/6)
  U_cdf=uniform.cdf(x,0,1)
  A_cdf=beta.cdf(x,0.5,0.5,loc=0,scale=1)
   
  #Empirical Distribution Function
  edf_=(np.arange(1,len(x)+1))/len(x)
  

  #KS Values   
  N_ksA.append(np.max(np.abs(N_cdf-edf_)))    
  U_ksA.append(np.max(np.abs(U_cdf-edf_)))    
  A_ksA.append(np.max(np.abs(A_cdf-edf_)))

 
  # PLOT
  plt.plot(x, N_cdf ,  label='truncated normal cdf ',color='blue') 
  plt.plot(x,U_cdf, label='Uniform cdf',color='red')
  plt.plot(x,A_cdf, label='Arcsine cdf',color='green')     
  plt.plot(x, edf_ ,  label='Edf ',color='black')   
  plt.legend()
  plt.title(("CDFs and EDF for n= " , samples))  
  plt.show() 
  
```

```python
plt.figure(figsize=(5, 5))

for trial in range (0,3):

  for i in range (2,6):
    
    Confusion_A(10**i,trial)
    

print(len(N_ksA))
print(N_ksA)
```

#CDF for Uniform Distribution

```python
N_ksU=[]
U_ksU=[]
A_ksU=[]
def Confusion_U(samples,trial):
  np.random.seed(10)
  x=uniform.rvs(0,1,size=samples)
  x.sort()
  N_cdf=truncnorm.cdf(x,a,b,loc=0.5,scale=1/6)
  U_cdf=uniform.cdf(x,0,1)
  A_cdf=beta.cdf(x,0.5,0.5,loc=0,scale=1)
   
  #Empirical Distribution Function
  edf_=(np.arange(1,len(x)+1))/len(x)
  

  #KS Values   
  N_ksU.append(np.max(np.abs(N_cdf-edf_)))    
  U_ksU.append(np.max(np.abs(U_cdf-edf_)))    
  A_ksU.append(np.max(np.abs(A_cdf-edf_)))

  
  # PLOT
  plt.plot(x, N_cdf ,  label='truncated normal cdf ',color='blue') 
  plt.plot(x,U_cdf, label='Uniform cdf',color='red')
  plt.plot(x,A_cdf, label='Arcsine cdf',color='green')     
  plt.plot(x, edf_ ,  label='Edf ',color='black')   
  plt.legend()
  plt.title(("CDFs and EDF for n= " , samples))  
  plt.show() 
```

```python
plt.figure(figsize=(5, 5))
for trial in range (0,3):
  for i in range (2,6):
    Confusion_U(10**i,trial)
    

print(len(N_ksU))
print(N_ksU)
```

# Taking Average of K-S Statistics over 3 Trials

```python
avgN_ksT=[]
avgU_ksT=[]
avgA_ksT=[]
avgN_ksU=[]
avgU_ksU=[]
avgA_ksU=[]
avgN_ksA=[]
avgU_ksA=[]
avgA_ksA=[]

for j in range(3+1):
  avgN_ksT.append(((N_ksT[j*3])+(N_ksT[j*3+1])+(N_ksT[j*3+2]))/3)
  avgU_ksT.append(((U_ksT[j*3])+(U_ksT[j*3+1])+(U_ksT[j*3+2]))/3)
  avgA_ksT.append(((A_ksT[j*3])+(A_ksT[j*3+1])+(A_ksT[j*3+2]))/3)
  avgN_ksU.append(((N_ksU[j*3])+(N_ksU[j*3+1])+(N_ksU[j*3+2]))/3)
  avgU_ksU.append(((U_ksU[j*3])+(U_ksU[j*3+1])+(U_ksU[j*3+2]))/3)
  avgA_ksU.append(((A_ksU[j*3])+(A_ksU[j*3+1])+(A_ksU[j*3+2]))/3)
  avgN_ksA.append(((N_ksA[j*3])+(N_ksA[j*3+1])+(N_ksA[j*3+2]))/3)
  avgU_ksA.append(((U_ksA[j*3])+(U_ksA[j*3+1])+(U_ksA[j*3+2]))/3)
  avgA_ksA.append(((A_ksA[j*3])+(A_ksA[j*3+1])+(A_ksA[j*3+2]))/3)
```
