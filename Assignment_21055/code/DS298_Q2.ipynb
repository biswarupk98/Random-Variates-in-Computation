```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import truncnorm
from scipy.stats import stats
# Parameters
a=-3
b=3
# Generate sample for Truncated Normal distribution 
samples=100
np.random.seed(10)
x=truncnorm.rvs(a,b,loc=0.5,scale=1/6,size=samples)
x.sort()
fx=truncnorm.pdf(x,a,b,loc=0.5,scale=1/6)
# Plotting
s_=[]
plt.figure(figsize=(15, 10))
for k in range(1, 9):
    # Transformation
    s = (1 + x)**(k * x)
    #dS/dX = kX * (1 + X)^(kX - 1) + (1 + X)^(kX) * ln(1 + X)
    dSdX = ((k*x) * (1 + x)**(k*x - 1)) + ((1 + x)**(k*x)) * np.log(1 + x)
    fs = fx * np.abs(1/ dSdX)
    s_.append(s)
    # Plot
    plt.plot(s, fs, label=f"k = {k}")
plt.xlabel("Career points:S")
plt.ylabel("The probability density:f_S(s)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i in range (0,8):
 plt.plot(x,s_[i],label=('S',i+1))
plt.title('S for various k vs random variable X')
plt.xlabel('sample:X')
plt.ylabel('S(k)')
plt.legend()
plt.show()
```
